```python
# Esta lista contiene todos los estados que pueden aparecer en la simulación.
# Cada estado representa una pantalla, botón, sección o resultado final
# dentro del recorrido del usuario.
estados = [
    "S0", "S1", "S2", "S3", "S4", "S5", "S6", "S7", "S8", "S9",
    "S10", "S11", "S12", "S13", "S14", "S15", "S16", "S17", "S18", "S19",
    "S20", "S21", "S22", "S23", "S24", "S25", "S26", "S27", "S28", "S29",
    "S30", "S31", "S32", "S33", "S34", "S35", "S36", "S37", "S38", "S39",
    "S40", "S41", "S42", "S43", "S44", "S45", "S46", "S47", "S48", "S49",
    "S50", "S51", "S52", "S53", "S54", "S55", "S56", "S57", "S58", "S59",
    "S60", "S61", "S62", "S63", "S64", "S65", "S66", "S67", "S68", "S69",
    "S70", "S71", "S72", "S73", "S74", "S75", "S76", "S77", "S78", "S79",
    "S80", "S81", "S82", "S83", "S84", "S85", "S86", "S87", "S88", "S89",
    "S90", "S91", "S92", "S93", "S94", "S95", "S96", "S97", "S98", "S99",
    "S100", "S101", "S102", "S103", "S104", "S105", "S106", "S107", "S108", "S109",
    "S110", "S111", "S112", "S113", "S114"
]
```

```python
# Se crea una matriz cuadrada donde las filas y columnas son los estados.
# Inicialmente todos los valores son 0.
#
# Cada fila representa el estado actual.
# Cada columna representa el estado siguiente.

matriz_conteos = pd.DataFrame(0, index=estados, columns=estados)

for recorrido in recorridos:
    for i in range(len(recorrido) - 1):
        estado_actual   = recorrido[i]
        estado_siguiente = recorrido[i + 1]
        matriz_conteos.loc[estado_actual, estado_siguiente] += 1

matriz_conteos
```

```python
# La matriz de conteos indica cuántas veces aparece cada transición.
# Para simular usuarios, necesitamos convertir esos conteos en probabilidades.
#
# Cada fila se divide entre la suma total de esa fila.
# Así, las probabilidades de salida desde cada estado suman 1.

matriz_probabilidades = matriz_conteos.div(
    matriz_conteos.sum(axis=1),
    axis=0
).fillna(0)

matriz_probabilidades
```

```python
# Esta función simula el recorrido de un usuario dentro de la plataforma.
def simular_usuario(matriz_probabilidades, estado_inicial='S1', max_pasos=50):
    estado_actual = estado_inicial
    recorrido = [estado_actual]

    estados_finales = [
        'S93', 'S94', 'S95', 'S96',
        'S97', 'S98', 'S99', 'S100', 'S101',
        'S102', 'S103', 'S104', 'S105', 'S106',
        'S107', 'S108', 'S109', 'S110',
        'S111', 'S112', 'S113', 'S114'
    ]

    for _ in range(max_pasos):
        if estado_actual in estados_finales:
            break

        if estado_actual not in matriz_probabilidades.index:
            break

        probabilidades = matriz_probabilidades.loc[estado_actual]

        if probabilidades.sum() == 0:
            break

        probabilidades = probabilidades / probabilidades.sum()

        siguiente_estado = np.random.choice(
            matriz_probabilidades.columns,
            p=probabilidades.values
        )

        recorrido.append(siguiente_estado)
        estado_actual = siguiente_estado

    return recorrido
```

```python

# Se simula el recorrido de un único usuario para revisar
# que la función esté funcionando correctamente.
recorrido_simulado = simular_usuario(matriz_probabilidades)


print(" → ".join(recorrido_simulado))
```

```python
# Se fija una semilla aleatoria para que los resultados sean reproducibles.
# Esto significa que, si se vuelve a ejecutar el código, se obtendrán
# los mismos resultados.

np.random.seed(42)

resultados = []

for i in range(10000):
    recorrido = simular_usuario(
        matriz_probabilidades,
        estado_inicial='S1',
        max_pasos=50
    )

    estado_final = recorrido[-1]

    resultados.append({
        "usuario": i + 1,
        "recorrido": " → ".join(recorrido),
        "estado_final": estado_final,
        "num_pasos": len(recorrido)
    })

df_resultados = pd.DataFrame(resultados)
df_resultados.head()
```
